```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from spicy import stats
```

```python
df=pd.read_csv('Auto.csv')
print(df)

x=df['mpg']
y=df['weight']
```

```python
plt.scatter(x,y)
plt.title("Collected Data")
plt.xlabel("mpg")
plt.ylabel("Weight")
plt.show()
```

```python
res = stats.linregress(x,y)
plt.plot(x,y,'.',label="Collected Data")
plt.plot(x,res.intercept+res.slope*x,'r',label="Data")
plt.xlabel("mpg")
plt.ylabel("Weight")
plt.legend()
```

```python
result =  stats.linregress(x,y)
print("Intercept :",result.intercept)
print("Slope :", result.slope)
```
